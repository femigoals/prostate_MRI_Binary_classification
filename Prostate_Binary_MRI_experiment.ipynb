# Inception V3

```python
import os
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_score, recall_score, f1_score, accuracy_score, roc_curve, precision_recall_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import gc
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, TensorBoard, CSVLogger
import SimpleITK as sitk
from IPython.display import display, HTML
import datetime
from itertools import cycle

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

def get_slice_dimension(img):
    """Determine which dimension represents slices."""
    dims = img.shape
    if len(dims) == 3:
        # Check which dimension looks like it could be slices (typically 15-100)
        for i, dim in enumerate(dims):
            if 15 <= dim <= 100:  # Typical slice count range
                return i
        # If no dimension in that range, return the smallest dimension
        return np.argmin(dims)
    return 2  # Default to last dimension

def extract_middle_slices(img, num_slices=13):
    """Extract 13 slices from the middle of the 3D volume."""
    slice_dim = get_slice_dimension(img)

    # Get total number of slices in the slice dimension
    if slice_dim == 0:
        total_slices = img.shape[0]
    elif slice_dim == 1:
        total_slices = img.shape[1]
    else:  # slice_dim == 2
        total_slices = img.shape[2]

    # If we have fewer slices than requested, take all slices
    if total_slices <= num_slices:
        slice_indices = list(range(total_slices))
    else:
        # Calculate start and end indices for middle slices
        start_idx = (total_slices - num_slices) // 2
        end_idx = start_idx + num_slices
        slice_indices = list(range(start_idx, end_idx))

    return slice_indices, slice_dim

def extract_slice(img, slice_idx, slice_dim):
    """Extract a slice from the 3D volume based on slice dimension."""
    if slice_dim == 0:
        return img[slice_idx, :, :]
    elif slice_dim == 1:
        return img[:, slice_idx, :]
    else:  # slice_dim == 2
        return img[:, :, slice_idx]

def load_mha_images(data_paths, labels, img_size=(224, 224), num_slices=13):
    """Load .mha files and extract middle slices."""
    images, processed_labels = [], []
    loaded_count = 0
    shape_info = []  # Track shapes for analysis

    print(f"\nLoading {len(data_paths)} MHA files...")
    print(f"Extracting {num_slices} middle slices from each file")

    for path, label in zip(data_paths, labels):
        try:
            # Read .mha file using SimpleITK
            img_sitk = sitk.ReadImage(path)
            img = sitk.GetArrayFromImage(img_sitk)  # Convert to numpy array

            # Store shape information
            shape_info.append(img.shape)

            # Print shape for debugging (first few files)
            if loaded_count < 3:
                print(f"File: {os.path.basename(path)}, Shape: {img.shape}")

            if img.ndim == 3:
                # Get middle slice indices and slice dimension
                slice_indices, slice_dim = extract_middle_slices(img, num_slices)

                for slice_idx in slice_indices:
                    # Extract slice based on slice dimension
                    slice_img = extract_slice(img, slice_idx, slice_dim)

                    # Normalize the slice
                    slice_min = slice_img.min()
                    slice_max = slice_img.max()
                    if slice_max > slice_min:  # Avoid division by zero
                        slice_img = (slice_img - slice_min) / (slice_max - slice_min)
                    else:
                        slice_img = np.zeros_like(slice_img)

                    # Resize to target dimensions
                    slice_img = cv2.resize(slice_img, img_size)

                    # Convert to 3-channel (RGB) for InceptionV3
                    slice_img_rgb = np.repeat(slice_img[:, :, np.newaxis], 3, axis=-1)

                    images.append(slice_img_rgb)
                    processed_labels.append(label)

                loaded_count += 1
                if loaded_count % 10 == 0:
                    print(f"  Loaded {loaded_count}/{len(data_paths)} files...")
            else:
                print(f"Unexpected dimensions {img.shape} in file: {path}")

            del img, img_sitk
            gc.collect()

        except Exception as e:
            print(f"Error loading {path}: {e}")

    # Print shape analysis
    if shape_info:
        shapes_array = np.array(shape_info)
        print(f"\nüìê Shape Analysis:")
        print(f"   Average shape: {np.mean(shapes_array, axis=0).astype(int)}")
        print(f"   Min shape: {np.min(shapes_array, axis=0)}")
        print(f"   Max shape: {np.max(shapes_array, axis=0)}")
        print(f"   Most common slice dimension: {np.argmax(np.bincount([get_slice_dimension(np.zeros(s)) for s in shape_info]))}")

    print(f"\n‚úì Successfully loaded {loaded_count}/{len(data_paths)} files")
    print(f"‚úì Total slices extracted: {len(images)}")
    print(f"‚úì Slices per file: {num_slices}")

    return np.array(images, dtype=np.float32), np.array(processed_labels, dtype=np.float32)

def get_data_paths_and_labels(base_path):
    """Get all .mha file paths and their corresponding labels."""
    data_paths, labels = [], []
    for label in ['normal', 'prostate_cancer']:
        folder = os.path.join(base_path, label)
        if os.path.exists(folder):
            for file in sorted(os.listdir(folder)):
                if file.endswith('.mha'):
                    data_paths.append(os.path.join(folder, file))
                    labels.append(label)

    print(f"\nüìä Dataset Summary:")
    print(f"   Total .mha files found: {len(data_paths)}")
    print(f"   Normal files: {labels.count('normal')}")
    print(f"   Prostate cancer files: {labels.count('prostate_cancer')}")

    # Convert labels to one-hot encoding
    label_binarizer = LabelBinarizer()
    labels_binary = label_binarizer.fit_transform(labels)
    labels_onehot = tf.keras.utils.to_categorical(labels_binary)

    return np.array(data_paths), labels_onehot, label_binarizer

def preview_dataset_images(data_paths, labels, num_samples=3, slices_to_show=5):
    """Preview sample images from the dataset."""
    fig = plt.figure(figsize=(15, 10))
    fig.suptitle(f'Dataset Sample Images (Showing {slices_to_show} middle slices per sample)',
                 fontsize=16, fontweight='bold', y=1.02)

    samples = {'prostate_cancer': 0, 'normal': 0}
    max_per_class = num_samples
    count = 1

    for idx, (path, label) in enumerate(zip(data_paths, labels)):
        class_name = 'prostate_cancer' if np.argmax(label) == 1 else 'normal'

        if samples[class_name] < max_per_class:
            try:
                img_sitk = sitk.ReadImage(path)
                img = sitk.GetArrayFromImage(img_sitk)

                if img.ndim == 3:
                    # Get middle slice indices (show only first few for preview)
                    slice_indices, slice_dim = extract_middle_slices(img, num_slices=slices_to_show)

                    for i, slice_idx in enumerate(slice_indices):
                        if samples[class_name] >= max_per_class:
                            break

                        slice_img = extract_slice(img, slice_idx, slice_dim)
                        slice_img = (slice_img - slice_img.min()) / (slice_img.max() - slice_img.min() + 1e-8)
                        slice_img = cv2.resize(slice_img, (224, 224))

                        # Calculate position in subplot grid
                        row = samples[class_name] * (slices_to_show + 1) + i + 1
                        col = 0 if class_name == 'normal' else slices_to_show + 2

                        ax = plt.subplot(max_per_class * (slices_to_show + 1),
                                         (slices_to_show + 2) * 2,
                                         count)
                        plt.imshow(slice_img, cmap='gray')
                        plt.title(f"Slice {slice_idx}", fontsize=8)
                        plt.axis('off')

                        # Add class label for first slice only
                        if i == 0:
                            plt.text(0.5, -0.1, class_name,
                                     fontsize=10, fontweight='bold',
                                     horizontalalignment='center',
                                     verticalalignment='top',
                                     transform=ax.transAxes,
                                     color='red' if class_name == 'prostate_cancer' else 'green')

                        # Add color border based on class
                        color = 'red' if class_name == 'prostate_cancer' else 'green'
                        for spine in ax.spines.values():
                            spine.set_edgecolor(color)
                            spine.set_linewidth(2)

                        count += 1

                    samples[class_name] += 1
            except Exception as e:
                print(f"Error previewing {path}: {e}")

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

def split_data(images, labels, test_size=0.2, val_size=0.2, random_state=42):
    """Split data into train, validation, and test sets."""
    # Get class labels for stratification
    class_labels = np.argmax(labels, axis=1)

    # First split: train+val vs test
    X_temp, X_test, y_temp, y_test = train_test_split(
        images, labels, test_size=test_size, stratify=class_labels, random_state=random_state
    )

    # Get new class labels for second split
    temp_class_labels = np.argmax(y_temp, axis=1)

    # Second split: train vs val
    val_relative_size = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_relative_size, stratify=temp_class_labels, random_state=random_state
    )

    print(f"\nüìà Data Split Summary:")
    print(f"   Training set: {X_train.shape[0]} slices")
    print(f"   Validation set: {X_val.shape[0]} slices")
    print(f"   Test set: {X_test.shape[0]} slices")

    # Plot class distribution
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    splits = [('Train', y_train), ('Validation', y_val), ('Test', y_test)]
    for idx, (split_name, split_labels) in enumerate(splits):
        class_dist = np.bincount(np.argmax(split_labels, axis=1))
        axes[idx].bar(['Normal', 'Cancer'], class_dist, color=['green', 'red'])
        axes[idx].set_title(f'{split_name} Set Distribution')
        axes[idx].set_ylabel('Number of Slices')
        axes[idx].tick_params(axis='x', rotation=45)
        for i, v in enumerate(class_dist):
            axes[idx].text(i, v, str(v), ha='center', va='bottom', fontweight='bold')

    plt.suptitle('Class Distribution Across Splits', fontsize=14, fontweight='bold', y=1.05)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    return X_train, X_val, X_test, y_train, y_val, y_test

def plot_roc_curve(y_true, y_pred_proba, class_names):
    """Plot ROC curve for binary classification."""
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba[:, 1])
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    display(plt.gcf())
    plt.close()

    return roc_auc

def plot_precision_recall_curve(y_true, y_pred_proba, class_names):
    """Plot Precision-Recall curve."""
    precision, recall, _ = precision_recall_curve(y_true, y_pred_proba[:, 1])
    pr_auc = auc(recall, precision)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='blue', lw=2, label=f'PR curve (AUC = {pr_auc:.2f})')
    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower left")
    plt.grid(True, alpha=0.3)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    display(plt.gcf())
    plt.close()

    return pr_auc

def evaluate_model(y_true, y_pred, y_pred_proba, class_names):
    """Evaluate model performance with various metrics."""
    cm = confusion_matrix(y_true, y_pred)

    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_pred_proba[:, 1])
    sensitivity = recall_score(y_true, y_pred)
    specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1]) if (cm[0, 0] + cm[0, 1]) > 0 else 0
    precision_val = precision_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    metrics = {
        'Accuracy': accuracy,
        'AUC Score': auc_score,
        'Sensitivity (Recall)': sensitivity,
        'Specificity': specificity,
        'Precision': precision_val,
        'F1 Score': f1
    }

    # Create a beautiful metrics table
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.axis('tight')
    ax.axis('off')

    table_data = []
    for metric, value in metrics.items():
        table_data.append([metric, f"{value:.4f}"])

    table = ax.table(cellText=table_data,
                     colLabels=['Metric', 'Value'],
                     cellLoc='center',
                     loc='center',
                     colColours=['#f0f0f0', '#f0f0f0'])

    table.auto_set_font_size(False)
    table.set_fontsize(11)
    table.scale(1.2, 1.5)

    for (row, col), cell in table.get_celld().items():
        if row == 0:
            cell.set_text_props(weight='bold', color='white')
            cell.set_facecolor('#2E86AB')
        else:
            if col == 1:
                cell.set_facecolor('#F5F5F5')

    plt.title('Model Performance Metrics', fontsize=14, fontweight='bold', pad=20)
    display(plt.gcf())
    plt.close()

    # Confusion matrix with annotations
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                annot_kws={"size": 14, "weight": "bold"})
    plt.title('Confusion Matrix', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    # Classification report
    print("\n" + "="*60)
    print("CLASSIFICATION REPORT")
    print("="*60)
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

    # Plot ROC and PR curves
    print("\n" + "="*60)
    print("PERFORMANCE CURVES")
    print("="*60)

    roc_auc = plot_roc_curve(y_true, y_pred_proba, class_names)
    pr_auc = plot_precision_recall_curve(y_true, y_pred_proba, class_names)

    # Add these to metrics
    metrics['ROC AUC'] = roc_auc
    metrics['PR AUC'] = pr_auc

    return metrics

def build_inception_model(input_shape=(224, 224, 3), num_classes=2):
    """Build InceptionV3 model for classification."""
    base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=input_shape)

    # Freeze early layers, fine-tune later layers
    for layer in base_model.layers[:-50]:
        layer.trainable = False
    for layer in base_model.layers[-50:]:
        layer.trainable = True

    # Add custom layers
    x = GlobalAveragePooling2D()(base_model.output)
    x = BatchNormalization()(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    output = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=output)

    # Use AdamW optimizer
    optimizer = tf.keras.optimizers.AdamW(
        learning_rate=1e-4,
        weight_decay=1e-4,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7
    )

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    return model

def plot_training_history(history):
    """Plot comprehensive training history."""
    # Determine how many metrics we have
    metrics_to_plot = []
    for key in ['accuracy', 'loss', 'auc', 'precision', 'recall']:
        if key in history.history:
            metrics_to_plot.append(key)

    num_metrics = len(metrics_to_plot)
    if num_metrics == 0:
        print("No training history to plot.")
        return

    # Create subplots
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    fig.suptitle('Training History', fontsize=16, fontweight='bold', y=1.02)

    metric_titles = {
        'accuracy': 'Accuracy',
        'loss': 'Loss',
        'auc': 'AUC',
        'precision': 'Precision',
        'recall': 'Recall'
    }

    for i, metric in enumerate(metrics_to_plot):
        if i >= len(axes):
            break

        ax = axes[i]
        train_metric = history.history.get(metric, [])
        val_metric = history.history.get(f'val_{metric}', [])

        if train_metric:
            ax.plot(train_metric, label='Train', linewidth=2)
        if val_metric:
            ax.plot(val_metric, label='Validation', linewidth=2)

        ax.set_title(metric_titles.get(metric, metric.capitalize()), fontsize=14, fontweight='bold')
        ax.set_xlabel('Epoch')
        ax.set_ylabel(metric_titles.get(metric, metric.capitalize()))
        ax.legend()
        ax.grid(True, alpha=0.3)

    # Plot learning rate if available
    if 'lr' in history.history and i+1 < len(axes):
        ax = axes[i+1]
        ax.plot(history.history['lr'], linewidth=2, color='purple')
        ax.set_title('Learning Rate', fontsize=14, fontweight='bold')
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Learning Rate')
        ax.set_yscale('log')
        ax.grid(True, alpha=0.3)
        i += 1

    # Hide unused subplots
    for j in range(i+1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

def main():
    """Main function to run the entire pipeline."""
    # Update BASE_PATH to your actual path
    BASE_PATH = '/content/drive/MyDrive/john_datasets/msc/dataset/Imaging'

    print("="*70)
    print("üß† PROSTATE CANCER CLASSIFICATION FROM MHA FILES")
    print("Using 13 middle slices from each 3D volume")
    print("="*70)

    # Create directories for saving outputs
    os.makedirs('models', exist_ok=True)
    os.makedirs('logs', exist_ok=True)

    # Get data paths and labels
    print("\n1Ô∏è‚É£  Scanning directory for MHA files...")
    data_paths, labels_onehot, label_binarizer = get_data_paths_and_labels(BASE_PATH)

    if len(data_paths) == 0:
        print("‚ùå No .mha files found! Please check your directory structure.")
        return

    # Load images - now taking 13 middle slices per file
    print("\n2Ô∏è‚É£  Loading and preprocessing MHA images...")
    images, preprocessed_labels = load_mha_images(data_paths, labels_onehot, num_slices=13)

    if len(images) == 0:
        print("‚ùå No images loaded! Check the error messages above.")
        return

    # Preview dataset - showing 5 middle slices per sample
    print("\n3Ô∏è‚É£  Previewing dataset images...")
    preview_dataset_images(data_paths[:4], labels_onehot[:4], num_samples=2, slices_to_show=5)

    # Split data
    print("\n4Ô∏è‚É£  Splitting data into train/val/test sets...")
    X_train, X_val, X_test, y_train, y_val, y_test = split_data(images, preprocessed_labels)

    # Build model
    print("\n5Ô∏è‚É£  Building InceptionV3 model...")
    model = build_inception_model()

    # Display model summary
    print("\nüìã Model Architecture Summary:")
    model.summary()

    # Create timestamp for logging
    timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

    # Define comprehensive callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True,
            verbose=1,
            min_delta=0.001,
            mode='min'
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=8,
            min_lr=1e-6,
            verbose=1,
            mode='min'
        ),
        ModelCheckpoint(
            f'models/best_model_{timestamp}.keras',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1,
            mode='max'
        ),
        ModelCheckpoint(
            f'models/best_model_auc_{timestamp}.keras',
            monitor='val_auc',
            save_best_only=True,
            verbose=1,
            mode='max'
        ),
        CSVLogger(f'logs/training_log_{timestamp}.csv'),
        TensorBoard(
            log_dir=f'logs/tensorboard_{timestamp}',
            histogram_freq=1,
            write_graph=True,
            write_images=True
        )
    ]

    # Compute class weights for imbalanced data
    print("\n6Ô∏è‚É£  Computing class weights...")
    class_labels = np.argmax(preprocessed_labels, axis=1)
    unique_classes, class_counts = np.unique(class_labels, return_counts=True)

    print(f"\nüìä Class distribution:")
    for cls, count in zip(label_binarizer.classes_, class_counts):
        print(f"   {cls}: {count} samples")

    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=unique_classes,
        y=class_labels
    )
    class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
    print(f"\n‚öñÔ∏è  Class weights: {class_weight_dict}")

    # Plot class weight visualization
    fig, ax = plt.subplots(figsize=(8, 4))
    bars = ax.bar(['Normal', 'Cancer'], class_counts, color=['green', 'red'])
    ax.set_title('Class Distribution and Weights', fontsize=14, fontweight='bold')
    ax.set_ylabel('Number of Slices')
    ax.set_xlabel('Class')

    for i, (bar, weight) in enumerate(zip(bars, class_weights)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'Weight: {weight:.2f}', ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    # Train model
    print("\n7Ô∏è‚É£  Training model...")
    print("-"*70)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        batch_size=32,  # Increased batch size due to more samples
        epochs=30,
        class_weight=class_weight_dict,
        callbacks=callbacks,
        verbose=1
    )

    # Plot training history
    print("\n8Ô∏è‚É£  Plotting training history...")
    plot_training_history(history)

    # Evaluate model
    print("\n9Ô∏è‚É£  Evaluating model on test set...")
    predictions = model.predict(X_test, verbose=0, batch_size=32)
    y_pred = np.argmax(predictions, axis=1)
    y_test_actual = np.argmax(y_test, axis=1)

    print("\n" + "="*70)
    print("üìä FINAL MODEL EVALUATION")
    print("="*70)

    metrics = evaluate_model(y_test_actual, y_pred, predictions, label_binarizer.classes_)

    # Save final model
    print("\nüîü  Saving model...")
    final_model_path = f'prostate_cancer_classifier_{timestamp}.keras'
    model.save(final_model_path)
    print(f"‚úÖ Model saved as '{final_model_path}'")

    # Print summary
    print("\n" + "="*70)
    print("üéØ TRAINING COMPLETE - SUMMARY")
    print("="*70)
    print(f"üìÅ Dataset: {len(data_paths)} MHA files")
    print(f"üñºÔ∏è  Total slices: {len(images)} (13 per file)")
    print(f"üèãÔ∏è  Training samples: {X_train.shape[0]}")
    print(f"‚öñÔ∏è  Validation samples: {X_val.shape[0]}")
    print(f"üß™ Test samples: {X_test.shape[0]}")
    print(f"üéØ Best Validation Accuracy: {max(history.history['val_accuracy']):.4f}")
    print(f"üéØ Best Validation AUC: {max(history.history['val_auc']):.4f}")
    print(f"üìä Test Accuracy: {metrics['Accuracy']:.4f}")
    print(f"üìä Test AUC: {metrics['AUC Score']:.4f}")
    print("="*70)

if __name__ == '__main__':
    # Install required packages if needed
    required_packages = ['SimpleITK', 'scipy', 'seaborn', 'opencv-python']

    for package in required_packages:
        try:
            if package == 'SimpleITK':
                import SimpleITK
            elif package == 'scipy':
                import scipy
            elif package == 'seaborn':
                import seaborn
            elif package == 'opencv-python':
                import cv2
        except ImportError:
            print(f"Installing {package}...")
            if package == 'opencv-python':
                !pip install opencv-python-headless
            else:
                !pip install {package}

    main()
```

```python
import os
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.applications import ResNet50  # Changed from InceptionV3 to ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_score, recall_score, f1_score, accuracy_score, roc_curve, precision_recall_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import gc
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, TensorBoard, CSVLogger
import SimpleITK as sitk
from IPython.display import display, HTML
import datetime

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

def get_slice_dimension(img):
    """Determine which dimension represents slices."""
    dims = img.shape
    if len(dims) == 3:
        # Check which dimension looks like it could be slices (typically 15-100)
        for i, dim in enumerate(dims):
            if 15 <= dim <= 100:  # Typical slice count range
                return i
        # If no dimension in that range, return the smallest dimension
        return np.argmin(dims)
    return 2  # Default to last dimension

def extract_middle_slices(img, num_slices=13):
    """Extract 13 slices from the middle of the 3D volume."""
    slice_dim = get_slice_dimension(img)

    # Get total number of slices in the slice dimension
    if slice_dim == 0:
        total_slices = img.shape[0]
    elif slice_dim == 1:
        total_slices = img.shape[1]
    else:  # slice_dim == 2
        total_slices = img.shape[2]

    # If we have fewer slices than requested, take all slices
    if total_slices <= num_slices:
        slice_indices = list(range(total_slices))
    else:
        # Calculate start and end indices for middle slices
        start_idx = (total_slices - num_slices) // 2
        end_idx = start_idx + num_slices
        slice_indices = list(range(start_idx, end_idx))

    return slice_indices, slice_dim

def extract_slice(img, slice_idx, slice_dim):
    """Extract a slice from the 3D volume based on slice dimension."""
    if slice_dim == 0:
        return img[slice_idx, :, :]
    elif slice_dim == 1:
        return img[:, slice_idx, :]
    else:  # slice_dim == 2
        return img[:, :, slice_idx]

def load_mha_images(data_paths, labels, img_size=(224, 224), num_slices=13):
    """Load .mha files and extract middle slices."""
    images, processed_labels = [], []
    loaded_count = 0
    shape_info = []  # Track shapes for analysis

    print(f"\nLoading {len(data_paths)} MHA files...")
    print(f"Extracting {num_slices} middle slices from each file")

    for path, label in zip(data_paths, labels):
        try:
            # Read .mha file using SimpleITK
            img_sitk = sitk.ReadImage(path)
            img = sitk.GetArrayFromImage(img_sitk)  # Convert to numpy array

            # Store shape information
            shape_info.append(img.shape)

            # Print shape for debugging (first few files)
            if loaded_count < 3:
                print(f"File: {os.path.basename(path)}, Shape: {img.shape}")

            if img.ndim == 3:
                # Get middle slice indices and slice dimension
                slice_indices, slice_dim = extract_middle_slices(img, num_slices)

                for slice_idx in slice_indices:
                    # Extract slice based on slice dimension
                    slice_img = extract_slice(img, slice_idx, slice_dim)

                    # Normalize the slice
                    slice_min = slice_img.min()
                    slice_max = slice_img.max()
                    if slice_max > slice_min:  # Avoid division by zero
                        slice_img = (slice_img - slice_min) / (slice_max - slice_min)
                    else:
                        slice_img = np.zeros_like(slice_img)

                    # Resize to target dimensions (224x224 for ResNet50)
                    slice_img = cv2.resize(slice_img, img_size)

                    # Convert to 3-channel (RGB) for ResNet50
                    slice_img_rgb = np.repeat(slice_img[:, :, np.newaxis], 3, axis=-1)

                    images.append(slice_img_rgb)
                    processed_labels.append(label)

                loaded_count += 1
                if loaded_count % 10 == 0:
                    print(f"  Loaded {loaded_count}/{len(data_paths)} files...")
            else:
                print(f"Unexpected dimensions {img.shape} in file: {path}")

            del img, img_sitk
            gc.collect()

        except Exception as e:
            print(f"Error loading {path}: {e}")

    # Print shape analysis
    if shape_info:
        shapes_array = np.array(shape_info)
        print(f"\nüìê Shape Analysis:")
        print(f"   Average shape: {np.mean(shapes_array, axis=0).astype(int)}")
        print(f"   Min shape: {np.min(shapes_array, axis=0)}")
        print(f"   Max shape: {np.max(shapes_array, axis=0)}")
        print(f"   Most common slice dimension: {np.argmax(np.bincount([get_slice_dimension(np.zeros(s)) for s in shape_info]))}")

    print(f"\n‚úì Successfully loaded {loaded_count}/{len(data_paths)} files")
    print(f"‚úì Total slices extracted: {len(images)}")
    print(f"‚úì Slices per file: {num_slices}")

    return np.array(images, dtype=np.float32), np.array(processed_labels, dtype=np.float32)

def get_data_paths_and_labels(base_path):
    """Get all .mha file paths and their corresponding labels."""
    data_paths, labels = [], []
    for label in ['normal', 'prostate_cancer']:
        folder = os.path.join(base_path, label)
        if os.path.exists(folder):
            for file in sorted(os.listdir(folder)):
                if file.endswith('.mha'):
                    data_paths.append(os.path.join(folder, file))
                    labels.append(label)

    print(f"\nüìä Dataset Summary:")
    print(f"   Total .mha files found: {len(data_paths)}")
    print(f"   Normal files: {labels.count('normal')}")
    print(f"   Prostate cancer files: {labels.count('prostate_cancer')}")

    # Convert labels to one-hot encoding
    label_binarizer = LabelBinarizer()
    labels_binary = label_binarizer.fit_transform(labels)
    labels_onehot = tf.keras.utils.to_categorical(labels_binary)

    return np.array(data_paths), labels_onehot, label_binarizer

def preview_dataset_images(data_paths, labels, num_samples=3, slices_to_show=5):
    """Preview sample images from the dataset."""
    fig = plt.figure(figsize=(15, 10))
    fig.suptitle(f'Dataset Sample Images (Showing {slices_to_show} middle slices per sample)',
                 fontsize=16, fontweight='bold', y=1.02)

    samples = {'prostate_cancer': 0, 'normal': 0}
    max_per_class = num_samples
    count = 1

    for idx, (path, label) in enumerate(zip(data_paths, labels)):
        class_name = 'prostate_cancer' if np.argmax(label) == 1 else 'normal'

        if samples[class_name] < max_per_class:
            try:
                img_sitk = sitk.ReadImage(path)
                img = sitk.GetArrayFromImage(img_sitk)

                if img.ndim == 3:
                    # Get middle slice indices (show only first few for preview)
                    slice_indices, slice_dim = extract_middle_slices(img, num_slices=slices_to_show)

                    for i, slice_idx in enumerate(slice_indices):
                        if samples[class_name] >= max_per_class:
                            break

                        slice_img = extract_slice(img, slice_idx, slice_dim)
                        slice_img = (slice_img - slice_img.min()) / (slice_img.max() - slice_img.min() + 1e-8)
                        slice_img = cv2.resize(slice_img, (224, 224))

                        ax = plt.subplot(max_per_class * (slices_to_show + 1),
                                         (slices_to_show + 2) * 2,
                                         count)
                        plt.imshow(slice_img, cmap='gray')
                        plt.title(f"Slice {slice_idx}", fontsize=8)
                        plt.axis('off')

                        # Add class label for first slice only
                        if i == 0:
                            plt.text(0.5, -0.1, class_name,
                                     fontsize=10, fontweight='bold',
                                     horizontalalignment='center',
                                     verticalalignment='top',
                                     transform=ax.transAxes,
                                     color='red' if class_name == 'prostate_cancer' else 'green')

                        # Add color border based on class
                        color = 'red' if class_name == 'prostate_cancer' else 'green'
                        for spine in ax.spines.values():
                            spine.set_edgecolor(color)
                            spine.set_linewidth(2)

                        count += 1

                    samples[class_name] += 1
            except Exception as e:
                print(f"Error previewing {path}: {e}")

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

def split_data(images, labels, test_size=0.2, val_size=0.2, random_state=42):
    """Split data into train, validation, and test sets."""
    # Get class labels for stratification
    class_labels = np.argmax(labels, axis=1)

    # First split: train+val vs test
    X_temp, X_test, y_temp, y_test = train_test_split(
        images, labels, test_size=test_size, stratify=class_labels, random_state=random_state
    )

    # Get new class labels for second split
    temp_class_labels = np.argmax(y_temp, axis=1)

    # Second split: train vs val
    val_relative_size = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_relative_size, stratify=temp_class_labels, random_state=random_state
    )

    print(f"\nüìà Data Split Summary:")
    print(f"   Training set: {X_train.shape[0]} slices")
    print(f"   Validation set: {X_val.shape[0]} slices")
    print(f"   Test set: {X_test.shape[0]} slices")

    # Plot class distribution
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    splits = [('Train', y_train), ('Validation', y_val), ('Test', y_test)]
    for idx, (split_name, split_labels) in enumerate(splits):
        class_dist = np.bincount(np.argmax(split_labels, axis=1))
        axes[idx].bar(['Normal', 'Cancer'], class_dist, color=['green', 'red'])
        axes[idx].set_title(f'{split_name} Set Distribution')
        axes[idx].set_ylabel('Number of Slices')
        axes[idx].tick_params(axis='x', rotation=45)
        for i, v in enumerate(class_dist):
            axes[idx].text(i, v, str(v), ha='center', va='bottom', fontweight='bold')

    plt.suptitle('Class Distribution Across Splits', fontsize=14, fontweight='bold', y=1.05)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    return X_train, X_val, X_test, y_train, y_val, y_test

def plot_roc_curve(y_true, y_pred_proba, class_names):
    """Plot ROC curve for binary classification."""
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba[:, 1])
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    display(plt.gcf())
    plt.close()

    return roc_auc

def plot_precision_recall_curve(y_true, y_pred_proba, class_names):
    """Plot Precision-Recall curve."""
    precision, recall, _ = precision_recall_curve(y_true, y_pred_proba[:, 1])
    pr_auc = auc(recall, precision)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='blue', lw=2, label=f'PR curve (AUC = {pr_auc:.2f})')
    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower left")
    plt.grid(True, alpha=0.3)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    display(plt.gcf())
    plt.close()

    return pr_auc

def evaluate_model(y_true, y_pred, y_pred_proba, class_names):
    """Evaluate model performance with various metrics."""
    cm = confusion_matrix(y_true, y_pred)

    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_pred_proba[:, 1])
    sensitivity = recall_score(y_true, y_pred)
    specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1]) if (cm[0, 0] + cm[0, 1]) > 0 else 0
    precision_val = precision_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    metrics = {
        'Accuracy': accuracy,
        'AUC Score': auc_score,
        'Sensitivity (Recall)': sensitivity,
        'Specificity': specificity,
        'Precision': precision_val,
        'F1 Score': f1
    }

    # Create a beautiful metrics table
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.axis('tight')
    ax.axis('off')

    table_data = []
    for metric, value in metrics.items():
        table_data.append([metric, f"{value:.4f}"])

    table = ax.table(cellText=table_data,
                     colLabels=['Metric', 'Value'],
                     cellLoc='center',
                     loc='center',
                     colColours=['#f0f0f0', '#f0f0f0'])

    table.auto_set_font_size(False)
    table.set_fontsize(11)
    table.scale(1.2, 1.5)

    for (row, col), cell in table.get_celld().items():
        if row == 0:
            cell.set_text_props(weight='bold', color='white')
            cell.set_facecolor('#2E86AB')
        else:
            if col == 1:
                cell.set_facecolor('#F5F5F5')

    plt.title('Model Performance Metrics', fontsize=14, fontweight='bold', pad=20)
    display(plt.gcf())
    plt.close()

    # Confusion matrix with annotations
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                annot_kws={"size": 14, "weight": "bold"})
    plt.title('Confusion Matrix', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    # Classification report
    print("\n" + "="*60)
    print("CLASSIFICATION REPORT")
    print("="*60)
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

    # Plot ROC and PR curves
    print("\n" + "="*60)
    print("PERFORMANCE CURVES")
    print("="*60)

    roc_auc = plot_roc_curve(y_true, y_pred_proba, class_names)
    pr_auc = plot_precision_recall_curve(y_true, y_pred_proba, class_names)

    # Add these to metrics
    metrics['ROC AUC'] = roc_auc
    metrics['PR AUC'] = pr_auc

    return metrics

def build_resnet50_model(input_shape=(224, 224, 3), num_classes=2):
    """Build ResNet50 model for classification."""
    # Load ResNet50 with ImageNet weights, excluding the top classification layer
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)

    # Freeze the base model layers first
    for layer in base_model.layers:
        layer.trainable = False

    # Unfreeze the last 50 layers for fine-tuning
    for layer in base_model.layers[-50:]:
        layer.trainable = True

    # Add custom classification head
    x = GlobalAveragePooling2D()(base_model.output)
    x = BatchNormalization()(x)
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    output = Dense(num_classes, activation='softmax')(x)

    # Create the model
    model = Model(inputs=base_model.input, outputs=output)

    # Compile the model with Adam optimizer
    optimizer = tf.keras.optimizers.Adam(
        learning_rate=1e-4,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7
    )

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    return model

def plot_training_history(history):
    """Plot comprehensive training history."""
    # Determine how many metrics we have
    metrics_to_plot = []
    for key in ['accuracy', 'loss', 'auc', 'precision', 'recall']:
        if key in history.history:
            metrics_to_plot.append(key)

    num_metrics = len(metrics_to_plot)
    if num_metrics == 0:
        print("No training history to plot.")
        return

    # Create subplots
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    fig.suptitle('Training History - ResNet50', fontsize=16, fontweight='bold', y=1.02)

    metric_titles = {
        'accuracy': 'Accuracy',
        'loss': 'Loss',
        'auc': 'AUC',
        'precision': 'Precision',
        'recall': 'Recall'
    }

    for i, metric in enumerate(metrics_to_plot):
        if i >= len(axes):
            break

        ax = axes[i]
        train_metric = history.history.get(metric, [])
        val_metric = history.history.get(f'val_{metric}', [])

        if train_metric:
            ax.plot(train_metric, label='Train', linewidth=2)
        if val_metric:
            ax.plot(val_metric, label='Validation', linewidth=2)

        ax.set_title(metric_titles.get(metric, metric.capitalize()), fontsize=14, fontweight='bold')
        ax.set_xlabel('Epoch')
        ax.set_ylabel(metric_titles.get(metric, metric.capitalize()))
        ax.legend()
        ax.grid(True, alpha=0.3)

    # Plot learning rate if available
    if 'lr' in history.history and i+1 < len(axes):
        ax = axes[i+1]
        ax.plot(history.history['lr'], linewidth=2, color='purple')
        ax.set_title('Learning Rate', fontsize=14, fontweight='bold')
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Learning Rate')
        ax.set_yscale('log')
        ax.grid(True, alpha=0.3)
        i += 1

    # Hide unused subplots
    for j in range(i+1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

def main():
    """Main function to run the entire pipeline."""
    # Update BASE_PATH to your actual path
    BASE_PATH = '/content/drive/MyDrive/john_datasets/msc/dataset/Imaging'

    print("="*70)
    print("üß† PROSTATE CANCER CLASSIFICATION FROM MHA FILES")
    print("Using ResNet50 with 13 middle slices from each 3D volume")
    print("="*70)

    # Create directories for saving outputs
    os.makedirs('models', exist_ok=True)
    os.makedirs('logs', exist_ok=True)

    # Get data paths and labels
    print("\n1Ô∏è‚É£  Scanning directory for MHA files...")
    data_paths, labels_onehot, label_binarizer = get_data_paths_and_labels(BASE_PATH)

    if len(data_paths) == 0:
        print("‚ùå No .mha files found! Please check your directory structure.")
        return

    # Load images - now taking 13 middle slices per file
    print("\n2Ô∏è‚É£  Loading and preprocessing MHA images...")
    images, preprocessed_labels = load_mha_images(data_paths, labels_onehot, num_slices=13)

    if len(images) == 0:
        print("‚ùå No images loaded! Check the error messages above.")
        return

    # Preview dataset - showing 5 middle slices per sample
    print("\n3Ô∏è‚É£  Previewing dataset images...")
    preview_dataset_images(data_paths[:4], labels_onehot[:4], num_samples=2, slices_to_show=5)

    # Split data
    print("\n4Ô∏è‚É£  Splitting data into train/val/test sets...")
    X_train, X_val, X_test, y_train, y_val, y_test = split_data(images, preprocessed_labels)

    # Build ResNet50 model
    print("\n5Ô∏è‚É£  Building ResNet50 model...")
    model = build_resnet50_model()

    # Display model summary
    print("\nüìã Model Architecture Summary (ResNet50):")
    model.summary()

    # Create timestamp for logging
    timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

    # Define comprehensive callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=20,  # Increased patience for ResNet50
            restore_best_weights=True,
            verbose=1,
            min_delta=0.001,
            mode='min'
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=10,  # Increased patience for learning rate reduction
            min_lr=1e-7,
            verbose=1,
            mode='min'
        ),
        ModelCheckpoint(
            f'models/best_resnet50_model_{timestamp}.keras',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1,
            mode='max'
        ),
        ModelCheckpoint(
            f'models/best_resnet50_model_auc_{timestamp}.keras',
            monitor='val_auc',
            save_best_only=True,
            verbose=1,
            mode='max'
        ),
        CSVLogger(f'logs/resnet50_training_log_{timestamp}.csv'),
        TensorBoard(
            log_dir=f'logs/resnet50_tensorboard_{timestamp}',
            histogram_freq=1,
            write_graph=True,
            write_images=True
        )
    ]

    # Compute class weights for imbalanced data
    print("\n6Ô∏è‚É£  Computing class weights...")
    class_labels = np.argmax(preprocessed_labels, axis=1)
    unique_classes, class_counts = np.unique(class_labels, return_counts=True)

    print(f"\nüìä Class distribution:")
    for cls, count in zip(label_binarizer.classes_, class_counts):
        print(f"   {cls}: {count} samples")

    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=unique_classes,
        y=class_labels
    )
    class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
    print(f"\n‚öñÔ∏è  Class weights: {class_weight_dict}")

    # Plot class weight visualization
    fig, ax = plt.subplots(figsize=(8, 4))
    bars = ax.bar(['Normal', 'Cancer'], class_counts, color=['green', 'red'])
    ax.set_title('Class Distribution and Weights - ResNet50', fontsize=14, fontweight='bold')
    ax.set_ylabel('Number of Slices')
    ax.set_xlabel('Class')

    for i, (bar, weight) in enumerate(zip(bars, class_weights)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'Weight: {weight:.2f}', ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    # Train model
    print("\n7Ô∏è‚É£  Training ResNet50 model...")
    print("-"*70)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        batch_size=32,
        epochs=30,  # Increased epochs for ResNet50
        class_weight=class_weight_dict,
        callbacks=callbacks,
        verbose=1
    )

    # Plot training history
    print("\n8Ô∏è‚É£  Plotting training history...")
    plot_training_history(history)

    # Evaluate model
    print("\n9Ô∏è‚É£  Evaluating model on test set...")
    predictions = model.predict(X_test, verbose=0, batch_size=32)
    y_pred = np.argmax(predictions, axis=1)
    y_test_actual = np.argmax(y_test, axis=1)

    print("\n" + "="*70)
    print("üìä FINAL MODEL EVALUATION - ResNet50")
    print("="*70)

    metrics = evaluate_model(y_test_actual, y_pred, predictions, label_binarizer.classes_)

    # Save final model
    print("\nüîü  Saving model...")
    final_model_path = f'prostate_cancer_resnet50_classifier_{timestamp}.keras'
    model.save(final_model_path)
    print(f"‚úÖ Model saved as '{final_model_path}'")

    # Print summary
    print("\n" + "="*70)
    print("üéØ TRAINING COMPLETE - SUMMARY")
    print("="*70)
    print(f"üìÅ Dataset: {len(data_paths)} MHA files")
    print(f"üñºÔ∏è  Total slices: {len(images)} (13 per file)")
    print(f"üèãÔ∏è  Training samples: {X_train.shape[0]}")
    print(f"‚öñÔ∏è  Validation samples: {X_val.shape[0]}")
    print(f"üß™ Test samples: {X_test.shape[0]}")
    print(f"ü§ñ Model: ResNet50")
    print(f"üéØ Best Validation Accuracy: {max(history.history['val_accuracy']):.4f}")
    print(f"üéØ Best Validation AUC: {max(history.history['val_auc']):.4f}")
    print(f"üìä Test Accuracy: {metrics['Accuracy']:.4f}")
    print(f"üìä Test AUC: {metrics['AUC Score']:.4f}")
    print(f"üéØ Test Sensitivity: {metrics['Sensitivity (Recall)']:.4f}")
    print(f"üéØ Test Specificity: {metrics['Specificity']:.4f}")
    print("="*70)

    # Create a final comparison visualization
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Training vs Validation Accuracy
    axes[0].plot(history.history['accuracy'], label='Train Accuracy', linewidth=2)
    axes[0].plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
    axes[0].set_title('ResNet50: Training vs Validation Accuracy', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Training vs Validation Loss
    axes[1].plot(history.history['loss'], label='Train Loss', linewidth=2)
    axes[1].plot(history.history['val_loss'], label='Val Loss', linewidth=2)
    axes[1].set_title('ResNet50: Training vs Validation Loss', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Loss')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.suptitle('ResNet50 Performance Summary', fontsize=16, fontweight='bold', y=1.05)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

if __name__ == '__main__':
    # Install required packages if needed
    required_packages = ['SimpleITK', 'scipy', 'seaborn', 'opencv-python']

    for package in required_packages:
        try:
            if package == 'SimpleITK':
                import SimpleITK
            elif package == 'scipy':
                import scipy
            elif package == 'seaborn':
                import seaborn
            elif package == 'opencv-python':
                import cv2
        except ImportError:
            print(f"Installing {package}...")
            if package == 'opencv-python':
                !pip install opencv-python-headless
            else:
                !pip install {package}

    main()
```

```python
import os
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_score, recall_score, f1_score, accuracy_score, roc_curve, precision_recall_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import gc
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, TensorBoard, CSVLogger
import SimpleITK as sitk
from IPython.display import display, HTML
import datetime

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

def get_slice_dimension(img):
    """Determine which dimension represents slices."""
    dims = img.shape
    if len(dims) == 3:
        # Check which dimension looks like it could be slices (typically 15-100)
        for i, dim in enumerate(dims):
            if 15 <= dim <= 100:  # Typical slice count range
                return i
        # If no dimension in that range, return the smallest dimension
        return np.argmin(dims)
    return 2  # Default to last dimension

def extract_middle_slices(img, num_slices=13):
    """Extract 13 slices from the middle of the 3D volume."""
    slice_dim = get_slice_dimension(img)

    # Get total number of slices in the slice dimension
    if slice_dim == 0:
        total_slices = img.shape[0]
    elif slice_dim == 1:
        total_slices = img.shape[1]
    else:  # slice_dim == 2
        total_slices = img.shape[2]

    # If we have fewer slices than requested, take all slices
    if total_slices <= num_slices:
        slice_indices = list(range(total_slices))
    else:
        # Calculate start and end indices for middle slices
        start_idx = (total_slices - num_slices) // 2
        end_idx = start_idx + num_slices
        slice_indices = list(range(start_idx, end_idx))

    return slice_indices, slice_dim

def extract_slice(img, slice_idx, slice_dim):
    """Extract a slice from the 3D volume based on slice dimension."""
    if slice_dim == 0:
        return img[slice_idx, :, :]
    elif slice_dim == 1:
        return img[:, slice_idx, :]
    else:  # slice_dim == 2
        return img[:, :, slice_idx]

def load_mha_images(data_paths, labels, img_size=(224, 224), num_slices=13):
    """Load .mha files and extract middle slices."""
    images, processed_labels = [], []
    loaded_count = 0
    shape_info = []  # Track shapes for analysis

    print(f"\nLoading {len(data_paths)} MHA files...")
    print(f"Extracting {num_slices} middle slices from each file")

    for path, label in zip(data_paths, labels):
        try:
            # Read .mha file using SimpleITK
            img_sitk = sitk.ReadImage(path)
            img = sitk.GetArrayFromImage(img_sitk)  # Convert to numpy array

            # Store shape information
            shape_info.append(img.shape)

            # Print shape for debugging (first few files)
            if loaded_count < 3:
                print(f"File: {os.path.basename(path)}, Shape: {img.shape}")

            if img.ndim == 3:
                # Get middle slice indices and slice dimension
                slice_indices, slice_dim = extract_middle_slices(img, num_slices)

                for slice_idx in slice_indices:
                    # Extract slice based on slice dimension
                    slice_img = extract_slice(img, slice_idx, slice_dim)

                    # Normalize the slice
                    slice_min = slice_img.min()
                    slice_max = slice_img.max()
                    if slice_max > slice_min:  # Avoid division by zero
                        slice_img = (slice_img - slice_min) / (slice_max - slice_min)
                    else:
                        slice_img = np.zeros_like(slice_img)

                    # Resize to target dimensions (224x224 for DenseNet121)
                    slice_img = cv2.resize(slice_img, img_size)

                    # Convert to 3-channel (RGB) for DenseNet121
                    slice_img_rgb = np.repeat(slice_img[:, :, np.newaxis], 3, axis=-1)

                    images.append(slice_img_rgb)
                    processed_labels.append(label)

                loaded_count += 1
                if loaded_count % 10 == 0:
                    print(f"  Loaded {loaded_count}/{len(data_paths)} files...")
            else:
                print(f"Unexpected dimensions {img.shape} in file: {path}")

            del img, img_sitk
            gc.collect()

        except Exception as e:
            print(f"Error loading {path}: {e}")

    # Print shape analysis
    if shape_info:
        shapes_array = np.array(shape_info)
        print(f"\nüìê Shape Analysis:")
        print(f"   Average shape: {np.mean(shapes_array, axis=0).astype(int)}")
        print(f"   Min shape: {np.min(shapes_array, axis=0)}")
        print(f"   Max shape: {np.max(shapes_array, axis=0)}")
        print(f"   Most common slice dimension: {np.argmax(np.bincount([get_slice_dimension(np.zeros(s)) for s in shape_info]))}")

    print(f"\n‚úì Successfully loaded {loaded_count}/{len(data_paths)} files")
    print(f"‚úì Total slices extracted: {len(images)}")
    print(f"‚úì Slices per file: {num_slices}")

    return np.array(images, dtype=np.float32), np.array(processed_labels, dtype=np.float32)

def get_data_paths_and_labels(base_path):
    """Get all .mha file paths and their corresponding labels."""
    data_paths, labels = [], []
    for label in ['normal', 'prostate_cancer']:
        folder = os.path.join(base_path, label)
        if os.path.exists(folder):
            for file in sorted(os.listdir(folder)):
                if file.endswith('.mha'):
                    data_paths.append(os.path.join(folder, file))
                    labels.append(label)

    print(f"\nüìä Dataset Summary:")
    print(f"   Total .mha files found: {len(data_paths)}")
    print(f"   Normal files: {labels.count('normal')}")
    print(f"   Prostate cancer files: {labels.count('prostate_cancer')}")

    # Convert labels to one-hot encoding
    label_binarizer = LabelBinarizer()
    labels_binary = label_binarizer.fit_transform(labels)
    labels_onehot = tf.keras.utils.to_categorical(labels_binary)

    return np.array(data_paths), labels_onehot, label_binarizer

def preview_dataset_images(data_paths, labels, num_samples=3, slices_to_show=5):
    """Preview sample images from the dataset."""
    fig = plt.figure(figsize=(15, 10))
    fig.suptitle(f'Dataset Sample Images (Showing {slices_to_show} middle slices per sample)',
                 fontsize=16, fontweight='bold', y=1.02)

    samples = {'prostate_cancer': 0, 'normal': 0}
    max_per_class = num_samples
    count = 1

    for idx, (path, label) in enumerate(zip(data_paths, labels)):
        class_name = 'prostate_cancer' if np.argmax(label) == 1 else 'normal'

        if samples[class_name] < max_per_class:
            try:
                img_sitk = sitk.ReadImage(path)
                img = sitk.GetArrayFromImage(img_sitk)

                if img.ndim == 3:
                    # Get middle slice indices (show only first few for preview)
                    slice_indices, slice_dim = extract_middle_slices(img, num_slices=slices_to_show)

                    for i, slice_idx in enumerate(slice_indices):
                        if samples[class_name] >= max_per_class:
                            break

                        slice_img = extract_slice(img, slice_idx, slice_dim)
                        slice_img = (slice_img - slice_img.min()) / (slice_img.max() - slice_img.min() + 1e-8)
                        slice_img = cv2.resize(slice_img, (224, 224))

                        ax = plt.subplot(max_per_class * (slices_to_show + 1),
                                         (slices_to_show + 2) * 2,
                                         count)
                        plt.imshow(slice_img, cmap='gray')
                        plt.title(f"Slice {slice_idx}", fontsize=8)
                        plt.axis('off')

                        # Add class label for first slice only
                        if i == 0:
                            plt.text(0.5, -0.1, class_name,
                                     fontsize=10, fontweight='bold',
                                     horizontalalignment='center',
                                     verticalalignment='top',
                                     transform=ax.transAxes,
                                     color='red' if class_name == 'prostate_cancer' else 'green')

                        # Add color border based on class
                        color = 'red' if class_name == 'prostate_cancer' else 'green'
                        for spine in ax.spines.values():
                            spine.set_edgecolor(color)
                            spine.set_linewidth(2)

                        count += 1

                    samples[class_name] += 1
            except Exception as e:
                print(f"Error previewing {path}: {e}")

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

def split_data(images, labels, test_size=0.2, val_size=0.2, random_state=42):
    """Split data into train, validation, and test sets."""
    # Get class labels for stratification
    class_labels = np.argmax(labels, axis=1)

    # First split: train+val vs test
    X_temp, X_test, y_temp, y_test = train_test_split(
        images, labels, test_size=test_size, stratify=class_labels, random_state=random_state
    )

    # Get new class labels for second split
    temp_class_labels = np.argmax(y_temp, axis=1)

    # Second split: train vs val
    val_relative_size = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_relative_size, stratify=temp_class_labels, random_state=random_state
    )

    print(f"\nüìà Data Split Summary:")
    print(f"   Training set: {X_train.shape[0]} slices")
    print(f"   Validation set: {X_val.shape[0]} slices")
    print(f"   Test set: {X_test.shape[0]} slices")

    # Plot class distribution
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    splits = [('Train', y_train), ('Validation', y_val), ('Test', y_test)]
    for idx, (split_name, split_labels) in enumerate(splits):
        class_dist = np.bincount(np.argmax(split_labels, axis=1))
        axes[idx].bar(['Normal', 'Cancer'], class_dist, color=['green', 'red'])
        axes[idx].set_title(f'{split_name} Set Distribution')
        axes[idx].set_ylabel('Number of Slices')
        axes[idx].tick_params(axis='x', rotation=45)
        for i, v in enumerate(class_dist):
            axes[idx].text(i, v, str(v), ha='center', va='bottom', fontweight='bold')

    plt.suptitle('Class Distribution Across Splits', fontsize=14, fontweight='bold', y=1.05)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    return X_train, X_val, X_test, y_train, y_val, y_test

def plot_roc_curve(y_true, y_pred_proba, class_names):
    """Plot ROC curve for binary classification."""
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba[:, 1])
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    display(plt.gcf())
    plt.close()

    return roc_auc

def plot_precision_recall_curve(y_true, y_pred_proba, class_names):
    """Plot Precision-Recall curve."""
    precision, recall, _ = precision_recall_curve(y_true, y_pred_proba[:, 1])
    pr_auc = auc(recall, precision)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='blue', lw=2, label=f'PR curve (AUC = {pr_auc:.2f})')
    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold')
    plt.legend(loc="lower left")
    plt.grid(True, alpha=0.3)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    display(plt.gcf())
    plt.close()

    return pr_auc

def evaluate_model(y_true, y_pred, y_pred_proba, class_names):
    """Evaluate model performance with various metrics."""
    cm = confusion_matrix(y_true, y_pred)

    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_pred_proba[:, 1])
    sensitivity = recall_score(y_true, y_pred)
    specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1]) if (cm[0, 0] + cm[0, 1]) > 0 else 0
    precision_val = precision_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    metrics = {
        'Accuracy': accuracy,
        'AUC Score': auc_score,
        'Sensitivity (Recall)': sensitivity,
        'Specificity': specificity,
        'Precision': precision_val,
        'F1 Score': f1
    }

    # Create a beautiful metrics table
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.axis('tight')
    ax.axis('off')

    table_data = []
    for metric, value in metrics.items():
        table_data.append([metric, f"{value:.4f}"])

    table = ax.table(cellText=table_data,
                     colLabels=['Metric', 'Value'],
                     cellLoc='center',
                     loc='center',
                     colColours=['#f0f0f0', '#f0f0f0'])

    table.auto_set_font_size(False)
    table.set_fontsize(11)
    table.scale(1.2, 1.5)

    for (row, col), cell in table.get_celld().items():
        if row == 0:
            cell.set_text_props(weight='bold', color='white')
            cell.set_facecolor('#2E86AB')
        else:
            if col == 1:
                cell.set_facecolor('#F5F5F5')

    plt.title('Model Performance Metrics', fontsize=14, fontweight='bold', pad=20)
    display(plt.gcf())
    plt.close()

    # Confusion matrix with annotations
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                annot_kws={"size": 14, "weight": "bold"})
    plt.title('Confusion Matrix', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    # Classification report
    print("\n" + "="*60)
    print("CLASSIFICATION REPORT")
    print("="*60)
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

    # Plot ROC and PR curves
    print("\n" + "="*60)
    print("PERFORMANCE CURVES")
    print("="*60)

    roc_auc = plot_roc_curve(y_true, y_pred_proba, class_names)
    pr_auc = plot_precision_recall_curve(y_true, y_pred_proba, class_names)

    # Add these to metrics
    metrics['ROC AUC'] = roc_auc
    metrics['PR AUC'] = pr_auc

    return metrics

def build_densenet121_model(input_shape=(224, 224, 3), num_classes=2):
    """Build DenseNet121 model for classification."""
    # Load DenseNet121 with ImageNet weights, excluding the top classification layer
    base_model = DenseNet121(
        weights='imagenet',
        include_top=False,
        input_shape=input_shape
    )

    # Freeze the base model layers initially
    for layer in base_model.layers:
        layer.trainable = False

    # Add custom classification head
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dropout(0.3)(x)
    output = Dense(num_classes, activation='softmax')(x)

    # Create the model
    model = Model(inputs=base_model.input, outputs=output)

    # Compile the model with Adam optimizer
    optimizer = tf.keras.optimizers.Adam(
        learning_rate=1e-4,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7
    )

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    return model

def unfreeze_densenet_layers(model, unfreeze_last_n=50):
    """Unfreeze the last N layers of DenseNet121 for fine-tuning."""
    # Get all layers in the model
    all_layers = model.layers

    # Find the DenseNet121 base model (it's at index 1 in our model)
    base_model = None
    for layer in all_layers:
        if isinstance(layer, tf.keras.Model):  # DenseNet121 is a Model
            base_model = layer
            break

    if base_model is None:
        print("‚ö†Ô∏è Could not find DenseNet121 base model in the model architecture.")
        return model

    # Count total layers in base model
    total_layers = len(base_model.layers)

    print(f"üîç Found DenseNet121 base model with {total_layers} layers")
    print(f"üîì Unfreezing last {unfreeze_last_n} layers...")

    # Freeze first layers, unfreeze last N layers
    for i, layer in enumerate(base_model.layers):
        if i < total_layers - unfreeze_last_n:
            layer.trainable = False
        else:
            layer.trainable = True

    # Also ensure our custom classification head is trainable
    for layer in model.layers:
        if layer not in [base_model]:
            layer.trainable = True

    # Count trainable parameters
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    non_trainable_params = sum([tf.keras.backend.count_params(w) for w in model.non_trainable_weights])
    total_params = trainable_params + non_trainable_params

    print(f"üìä Parameters: {trainable_params:,} trainable, {non_trainable_params:,} non-trainable")
    print(f"üìä Total: {total_params:,} parameters")

    # Recompile the model with lower learning rate for fine-tuning
    optimizer = tf.keras.optimizers.Adam(
        learning_rate=1e-5,  # Lower learning rate for fine-tuning
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7
    )

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    return model

def plot_training_history(history):
    """Plot comprehensive training history."""
    # Determine how many metrics we have
    metrics_to_plot = []
    for key in ['accuracy', 'loss', 'auc', 'precision', 'recall']:
        if key in history.history:
            metrics_to_plot.append(key)

    num_metrics = len(metrics_to_plot)
    if num_metrics == 0:
        print("No training history to plot.")
        return

    # Create subplots
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    fig.suptitle('Training History - DenseNet121', fontsize=16, fontweight='bold', y=1.02)

    metric_titles = {
        'accuracy': 'Accuracy',
        'loss': 'Loss',
        'auc': 'AUC',
        'precision': 'Precision',
        'recall': 'Recall'
    }

    for i, metric in enumerate(metrics_to_plot):
        if i >= len(axes):
            break

        ax = axes[i]
        train_metric = history.history.get(metric, [])
        val_metric = history.history.get(f'val_{metric}', [])

        if train_metric:
            ax.plot(train_metric, label='Train', linewidth=2)
        if val_metric:
            ax.plot(val_metric, label='Validation', linewidth=2)

        ax.set_title(metric_titles.get(metric, metric.capitalize()), fontsize=14, fontweight='bold')
        ax.set_xlabel('Epoch')
        ax.set_ylabel(metric_titles.get(metric, metric.capitalize()))
        ax.legend()
        ax.grid(True, alpha=0.3)

    # Plot learning rate if available
    if 'lr' in history.history and i+1 < len(axes):
        ax = axes[i+1]
        ax.plot(history.history['lr'], linewidth=2, color='purple')
        ax.set_title('Learning Rate', fontsize=14, fontweight='bold')
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Learning Rate')
        ax.set_yscale('log')
        ax.grid(True, alpha=0.3)
        i += 1

    # Hide unused subplots
    for j in range(i+1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

def main():
    """Main function to run the entire pipeline."""
    # Update BASE_PATH to your actual path
    BASE_PATH = '/content/drive/MyDrive/john_datasets/msc/dataset/Imaging'

    print("="*70)
    print("üß† PROSTATE CANCER CLASSIFICATION FROM MHA FILES")
    print("Using DenseNet121 with 13 middle slices from each 3D volume")
    print("="*70)

    # Create directories for saving outputs
    os.makedirs('models', exist_ok=True)
    os.makedirs('logs', exist_ok=True)

    # Get data paths and labels
    print("\n1Ô∏è‚É£  Scanning directory for MHA files...")
    data_paths, labels_onehot, label_binarizer = get_data_paths_and_labels(BASE_PATH)

    if len(data_paths) == 0:
        print("‚ùå No .mha files found! Please check your directory structure.")
        return

    # Load images - now taking 13 middle slices per file
    print("\n2Ô∏è‚É£  Loading and preprocessing MHA images...")
    images, preprocessed_labels = load_mha_images(data_paths, labels_onehot, num_slices=13)

    if len(images) == 0:
        print("‚ùå No images loaded! Check the error messages above.")
        return

    # Preview dataset - showing 5 middle slices per sample
    print("\n3Ô∏è‚É£  Previewing dataset images...")
    preview_dataset_images(data_paths[:4], labels_onehot[:4], num_samples=2, slices_to_show=5)

    # Split data
    print("\n4Ô∏è‚É£  Splitting data into train/val/test sets...")
    X_train, X_val, X_test, y_train, y_val, y_test = split_data(images, preprocessed_labels)

    # Build DenseNet121 model
    print("\n5Ô∏è‚É£  Building DenseNet121 model...")
    model = build_densenet121_model()

    # Display model summary
    print("\nüìã Model Architecture Summary (DenseNet121):")
    model.summary()

    # Create timestamp for logging
    timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

    # Define comprehensive callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=20,  # Good patience for DenseNet
            restore_best_weights=True,
            verbose=1,
            min_delta=0.001,
            mode='min'
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=10,  # Good patience for learning rate reduction
            min_lr=1e-7,
            verbose=1,
            mode='min'
        ),
        ModelCheckpoint(
            f'models/best_densenet121_model_{timestamp}.keras',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1,
            mode='max'
        ),
        ModelCheckpoint(
            f'models/best_densenet121_model_auc_{timestamp}.keras',
            monitor='val_auc',
            save_best_only=True,
            verbose=1,
            mode='max'
        ),
        CSVLogger(f'logs/densenet121_training_log_{timestamp}.csv'),
        TensorBoard(
            log_dir=f'logs/densenet121_tensorboard_{timestamp}',
            histogram_freq=1,
            write_graph=True,
            write_images=True
        )
    ]

    # Compute class weights for imbalanced data
    print("\n6Ô∏è‚É£  Computing class weights...")
    class_labels = np.argmax(preprocessed_labels, axis=1)
    unique_classes, class_counts = np.unique(class_labels, return_counts=True)

    print(f"\nüìä Class distribution:")
    for cls, count in zip(label_binarizer.classes_, class_counts):
        print(f"   {cls}: {count} samples")

    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=unique_classes,
        y=class_labels
    )
    class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
    print(f"\n‚öñÔ∏è  Class weights: {class_weight_dict}")

    # Plot class weight visualization
    fig, ax = plt.subplots(figsize=(8, 4))
    bars = ax.bar(['Normal', 'Cancer'], class_counts, color=['green', 'red'])
    ax.set_title('Class Distribution and Weights - DenseNet121', fontsize=14, fontweight='bold')
    ax.set_ylabel('Number of Slices')
    ax.set_xlabel('Class')

    for i, (bar, weight) in enumerate(zip(bars, class_weights)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'Weight: {weight:.2f}', ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    display(plt.gcf())
    plt.close()

    # Train with frozen base model
    print("\n7Ô∏è‚É£  Training DenseNet121 with frozen base layers...")
    print("-"*70)

    initial_epochs = 20
    history_initial = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        batch_size=32,
        epochs=initial_epochs,
        class_weight=class_weight_dict,
        callbacks=callbacks,
        verbose=1
    )

    # Fine-tune with unfrozen layers
    print("\n8Ô∏è‚É£  Fine-tuning DenseNet121 (unfreezing last 50 layers)...")
    print("-"*70)

    # Unfreeze last 50 layers for fine-tuning
    model = unfreeze_densenet_layers(model, unfreeze_last_n=50)

    # Update callbacks for fine-tuning
    fine_tune_callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True,
            verbose=1,
            min_delta=0.001,
            mode='min'
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=8,
            min_lr=1e-7,
            verbose=1,
            mode='min'
        ),
        ModelCheckpoint(
            f'models/fine_tuned_densenet121_{timestamp}.keras',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1,
            mode='max'
        )
    ]

    # Fine-tune the model
    fine_tune_epochs = 20
    total_epochs = initial_epochs + fine_tune_epochs

    history_fine_tune = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        initial_epoch=history_initial.epoch[-1] + 1 if history_initial.epoch else 0,
        epochs=total_epochs,
        class_weight=class_weight_dict,
        callbacks=fine_tune_callbacks,
        verbose=1
    )

    # Combine histories
    combined_history = {}
    for key in history_initial.history.keys():
        combined_history[key] = history_initial.history[key] + history_fine_tune.history[key]

    # Plot training history
    print("\n9Ô∏è‚É£  Plotting combined training history...")
    plot_training_history(combined_history)

    # Evaluate model
    print("\nüîü  Evaluating model on test set...")
    predictions = model.predict(X_test, verbose=0, batch_size=32)
    y_pred = np.argmax(predictions, axis=1)
    y_test_actual = np.argmax(y_test, axis=1)

    print("\n" + "="*70)
    print("üìä FINAL MODEL EVALUATION - DenseNet121")
    print("="*70)

    metrics = evaluate_model(y_test_actual, y_pred, predictions, label_binarizer.classes_)

    # Save final model
    print("\nüíæ Saving final model...")
    final_model_path = f'prostate_cancer_densenet121_classifier_{timestamp}.keras'
    model.save(final_model_path)
    print(f"‚úÖ Model saved as '{final_model_path}'")

    # Print summary
    print("\n" + "="*70)
    print("üéØ TRAINING COMPLETE - SUMMARY")
    print("="*70)
    print(f"üìÅ Dataset: {len(data_paths)} MHA files")
    print(f"üñºÔ∏è  Total slices: {len(images)} (13 per file)")
    print(f"üèãÔ∏è  Training samples: {X_train.shape[0]}")
    print(f"‚öñÔ∏è  Validation samples: {X_val.shape[0]}")
    print(f"üß™ Test samples: {X_test.shape[0]}")
    print(f"ü§ñ Model: DenseNet121")
    print(f"üîß Training strategy: 2-stage (frozen base + fine-tuning)")
    print(f"üéØ Best Validation Accuracy: {max(combined_history['val_accuracy']):.4f}")
    print(f"üéØ Best Validation AUC: {max(combined_history['val_auc']):.4f}")
    print(f"üìä Test Accuracy: {metrics['Accuracy']:.4f}")
    print(f"üìä Test AUC: {metrics['AUC Score']:.4f}")
    print(f"üéØ Test Sensitivity: {metrics['Sensitivity (Recall)']:.4f}")
    print(f"üéØ Test Specificity: {metrics['Specificity']:.4f}")
    print(f"üìà Test F1 Score: {metrics['F1 Score']:.4f}")
    print("="*70)

    # Create a final comparison visualization
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Training vs Validation Accuracy
    axes[0].plot(combined_history['accuracy'], label='Train Accuracy', linewidth=2)
    axes[0].plot(combined_history['val_accuracy'], label='Val Accuracy', linewidth=2)
    if len(combined_history['accuracy']) > initial_epochs:
        axes[0].axvline(x=initial_epochs-1, color='r', linestyle='--', alpha=0.7, label='Fine-tuning start')
    axes[0].set_title('DenseNet121: Training vs Validation Accuracy', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Training vs Validation Loss
    axes[1].plot(combined_history['loss'], label='Train Loss', linewidth=2)
    axes[1].plot(combined_history['val_loss'], label='Val Loss', linewidth=2)
    if len(combined_history['loss']) > initial_epochs:
        axes[1].axvline(x=initial_epochs-1, color='r', linestyle='--', alpha=0.7, label='Fine-tuning start')
    axes[1].set_title('DenseNet121: Training vs Validation Loss', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Loss')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.suptitle('DenseNet121 Performance Summary', fontsize=16, fontweight='bold', y=1.05)
    plt.tight_layout()
    display(plt.gcf())
    plt.close()

if __name__ == '__main__':
    # Install required packages if needed
    required_packages = ['SimpleITK', 'scipy', 'seaborn', 'opencv-python']

    for package in required_packages:
        try:
            if package == 'SimpleITK':
                import SimpleITK
            elif package == 'scipy':
                import scipy
            elif package == 'seaborn':
                import seaborn
            elif package == 'opencv-python':
                import cv2
        except ImportError:
            print(f"Installing {package}...")
            if package == 'opencv-python':
                !pip install opencv-python-headless
            else:
                !pip install {package}

    main()
```
